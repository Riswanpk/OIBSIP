Task 1 : IRIS FLOWER CLASSIFICATION

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 1 : Importing  The Dataset

```python
iris=pd.read_csv("Iris.csv")
iris

```

```python
iris=iris.drop(columns=['Id'])
```

Step 2 : Separating features and target variables

```python
x=iris.iloc[:,:-1].values
y=iris.iloc[:,-1].values
```

Step 3 : Splitting the dataset into training and testing set

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Step 4 : Standardizing the features

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

Step 5 : Training a KNearest Neighbors Classifier

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

Step 6 : Making Prediction on the test set

```python
y_pred=knn.predict(x_test)
```

Step 7 : Evaluating accuracy of the model

```python
accuracy=accuracy_score(y_test,y_pred)*100
print(f"Accuracy: {accuracy:.2f}%")
```

Step 8 : Pairplot to visualize relationship between features

```python
sns.pairplot(iris,hue='Species')
plt.title("pair plot of iris dataset")
plt.show()
```
