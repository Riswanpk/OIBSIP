### TASK 02 : UNEMPLOYMENT ANALYSIS WITH PYTHON

**Step 01 : Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

**Step 02 : Loading the Dataset**

```python
data=pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
print(data.head())
```

**Step 03 : Reading the Dataset**

```python
data.info
```

**Step 04 : Check for Missing Values**

```python
print(data.isnull().sum())
```

**Step 05 : Renaming all the Columns**

```python
data.columns=['States','Date','Frequency','Estimated Unemployment Rate',
              'Estimated Employed','Estimated Labour Participation Rate',
              'Region','longitude','latitude']
```

```python
print(data)
```

**Step 06 : Visualizing the Correlation b/w the features of the Dataset**

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10,10))
sns.heatmap(data.corr())
plt.show()
```

Histogram for Estimated number of employees according to different Regions of India

```python
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

Histogram for the Unemployment Rate according to different Regions of India

```python
plt.figure(figsize=(10,7))
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Unemployment Rate',hue='Region',data=data)
plt.show()
```

Creating a Dashboard to analyze the Unemployment rate of each Indian state by region (Using sunburst plot)

```python
unemployment=data[['States','Region','Estimated Unemployment Rate']]
figure=px.sunburst(unemployment,path=['Region','States'],
                   values='Estimated Unemployment Rate',
                   width=700,height=700,color_continuous_scale='RdBeGn',
                   title='Unemployment Rate in India')
figure.show()
```
